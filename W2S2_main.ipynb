```python
from google.colab import userdata
from openai import OpenAI
# Load key from Google Colab Secrets
api_key = userdata.get('OPENAI_API_KEY')
client = OpenAI(
    # This is the default and can be omitted
    api_key=api_key,
)

# Ensure 'df' is defined, assuming df_org is available from previous cells
df = df_org.copy()

# Convert the first few rows to a string to send to OpenAI
data_preview = df.head(10).to_csv(index=False)
print(data_preview)

instructions_content = """You are an expert data scientist with extensive knowledge of predictive analysis and linear regression."""

user_message_content = f"""Dataset: Laptop classification\nHere are the first 10 rows of the dataset:\n{data_preview}\n\n
1. Describe the dataset structure in text\nWithout using any visualisation, summarise the dataset in plain text so an external AI can understand its\nscale and layout.\n
Include key descriptors such as dataset size, column types, missing values, and anything unusual. You\nshould draw from the checks you attempted in the previous session’s
“Additional Exercise: With\nWindsurf”. If you have not completed that exercise, attempt it now before continuing.\n

2. Decide what context the AI should know first\nThink about what an AI model needs to know before giving advice.\nThis may include what one row represents,
what the target is, whether the data is clean, and any known\nconstraints or assumptions.\n

3. Combine everything into a single API payload\n
Using the OpenAI API code shown in the previous slide, combine your dataset description and context into\n
one clear text payload.\nWrite it as if it will be sent directly to an AI model, with no extra explanation or screenshots\n

4. include python codes"""

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": instructions_content},
        {"role": "user", "content": user_message_content}
    ],
)
print(response.choices[0].message.content)
```
